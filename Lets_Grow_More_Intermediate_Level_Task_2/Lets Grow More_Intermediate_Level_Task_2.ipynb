**Step 6 : Predictions using Decision Tree Algorithm**

```python
data = data.drop('Id', axis=1)
data.head()
```

```python
x = data.iloc[:, 0:4]
x.head()
```

```python
y = (data.iloc[:, 4])
y.head().to_frame()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

std = StandardScaler()
x = std.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

**Step 7 : Evaluate the Machine Learning Model**

```python
from sklearn.metrics import accuracy_score, confusion_matrix
acc = accuracy_score(y_test, y_pred)
print("Accuracy of the model : ", str(acc*100) + "%")
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
lst = data['Species'].unique().tolist()
df_cm = pd.DataFrame(data = cm, index = lst, columns = lst)
df_cm
```
