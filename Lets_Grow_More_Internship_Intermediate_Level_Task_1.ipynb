##### Step 2 : Importing the libraries

```python
data = pd.read_csv("globalterrorismdb_0718dist.csv", encoding ='latin1')
data.head()
```

##### Step 5 : Filtering the required columns for Analysis

```python
data=data[['eventid', 'iyear', 'imonth', 'country', 'region','provstate','city','crit1', 'crit2', 'crit3','success', 'suicide', 'attacktype1','targtype1','natlty1','gname','guncertain1','claimed','weaptype1','nkill','nwound']]
data.head()
```

##### Step 6 : Handling the null values

```python
data.isnull().sum()
```

```python
data['nkill'] = data['nkill'].fillna(0)
data['nwound'] = data['nwound'].fillna(0)
data['casualities'] = data['nkill'] + data['nwound']
data.isnull().sum()
```

```python
data.describe()
```

```python
n_unique_countries = data.country.nunique()
n_unique_regions = data.region.nunique()
n_years_covered = data.claimed.nunique()
min_year = data.iyear.min()
max_year = data.iyear.max()
n_terrorist_attacks = data.index.nunique()
n_casualities = data.casualities.sum()
n_kills = data.nkill.sum()
n_wounded = data.nwound.sum()
```

```python
print(f"There are {n_unique_countries} countries from {n_unique_regions} different regions covered in this terrorism dataset.")
print(f"The data spans starting from year {min_year} to {max_year} and the claims are made for {n_years_covered} years.")
print(f"The total number of terrorist attacks which took place within this timespan is {n_terrorist_attacks} attacks resulting in {n_casualities} casualities of which {n_kills} are kills and {n_wounded} are wounds")
```

**Step 7** : Data Visualization

```python
plt.subplots(figsize = (16, 9))
sns.countplot(x='iyear', data=data, edgecolor=sns.color_palette('dark', 10), palette='RdYlGn_r')
plt.xticks(rotation=90)
plt.title("Number of Terrorist Activities by year")
plt.show()
```

```python
yearc=data[['iyear','casualities']].groupby('iyear').sum()
yearc.plot(kind='bar',color='red',figsize=(15,6))
plt.title("Number of Casualities by year")
plt.xlabel('Years',fontsize=15)
plt.ylabel('No. of Casualities',fontsize=15)
plt.show()
```

```python
data['attacktype1'].value_counts().plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation=50)
plt.xlabel('Attack Type',fontsize=20)
plt.ylabel('Number of attacks')
plt.title('Number of attacks by attack type')
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(x='targtype1', data=data, palette='gist_heat', edgecolor=sns.color_palette("crest"));
plt.xticks(rotation=90)
plt.xlabel('Attack type',fontsize=20)
plt.ylabel('count')
plt.title('Number of Attacks by Type of Attacks')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
country_attacks = data.gname.value_counts()[:15].reset_index()
country_attacks.columns = ['gname', 'Total Attacks']
sns.barplot(x = country_attacks.gname, y = country_attacks['Total Attacks'], palette='OrRd_r', edgecolor=sns.color_palette('dark', 10))
plt.xticks(rotation = 90)
plt.title("Number of Total Attacks in Each Country by Terrorist Organizations")
plt.show()
```
