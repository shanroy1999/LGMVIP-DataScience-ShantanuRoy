#### TASK 1 - Handwriting Recognition Neural Network

##### Step 1 : Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

##### Step 2 : Loading Dataset

```python
mnist= tf.keras.datasets.mnist
```

##### Step 3 : Splitting Dataset into Training and Testing data

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

##### Step 4 : Data Exploration and Visualization

```python
fig=plt.figure(figsize=(15,3))
for i in range(20):
  ax=fig.add_subplot(2,10,i+1)
  ax.imshow(np.squeeze(x_train[i]),cmap='Reds')
  ax.set_title(y_train[i])
```

```python
print(x_train.shape)
print(x_train[0])
```

##### Step 5: Normalizing the matrix array of number images

```python
xtrain = x_train/255.0
xtest = x_test/255.0
```

##### Step 6: Flattening the 2-dimensional array into one dimensional array or a single column

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)),
                                    tf.keras.layers.Dense(128, activation ='relu'),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
model.summary()
```

##### Step 7 : Using AdamOptimizer for compiling the model

```python
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss ='sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

##### Step 8 : Fitting of the training data into the model

```python
model.fit(xtrain,y_train, epochs=5)
```

##### For 5 iterations => Accuracy = 98.67%

##### Fitting of the training data into the model for 9 iterations

```python
model.fit(xtrain,y_train, epochs=9)
```

##### For 9 iterations => Accuracy = 99.78%

##### Step 9 : Evaluating the model against the test set

```python
print(model.evaluate(x_test,y_test))
```

##### 97% of total accuracy of our neural network has been achieved

##### Fitting the model for 10 epochs and batch size 32 with validation split 0.1

```python
history=model.fit(x_train,y_train,epochs=10,batch_size=32,validation_split=0.1)
```

#### Step 10 : Plotting Accuracy of Model

```python
plt.title("Accuracy and Loss")
plt.xlabel("Epoch")
plt.ylabel("acc/Loss")
plt.plot(history.history['accuracy'],label='acc')
plt.plot(history.history["val_accuracy"],label='val')
plt.show()
```

```python
plt.title("Accuracy and Loss")
plt.xlabel("Epoch")
plt.ylabel("acc/Loss")
plt.plot(history.history['loss'],label='acc')
plt.plot(history.history["val_loss"],label='val')
plt.show()
```
