**Step 2 : Importing the OpenCV Library**

```python
import cv2
```

**Step 3 : Reading and displaying the image**

```python
house_image = cv2.imread("abc.jpg")
cv2.imshow("Amazing House", house_image)
cv2.waitKey(0)
```

Step 4 : Converting the image to Grayscale and display**

```python
house_image_gray = cv2.cvtColor(house_image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale House", house_image_gray)
cv2.waitKey(0)
```

Step 5 : Inverting the Grayscale Image

```python
inverted_house_image = 255 - house_image_gray
cv2.imshow("House Grayscale Inverted", inverted_house_image)
cv2.waitKey(0)
```

Step 6 : Blurring the Grayscale image using GaussianBlur

```python
house_image_blurred = cv2.GaussianBlur(inverted_house_image, (21, 21), 0)
cv2.imshow("House Grayscale Inverted and Blurred", house_image_blurred)
cv2.waitKey(0)
```

Step 7 : Inverting the blurred grayscale image to generate the Sketch

```python
inverted_house_image_blurred = 255 - house_image_blurred
house_pencil_sketch = cv2.divide(house_image_gray, inverted_house_image_blurred, scale=256.0)
cv2.imshow("House Sketch", house_pencil_sketch)
cv2.waitKey(0)
```
